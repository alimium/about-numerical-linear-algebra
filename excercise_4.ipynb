```python
import numpy as np
import matplotlib.pyplot as plt
from tools.generators import hilbert, vandermonde
from tools.definitions import condition_number as cond
```

```python
# Generate Hilbert matrix
n = 25
conditions = []
for i in range(1, n + 1):
    H = hilbert(i)
    conditions.append(cond(H))
    print(f"n = {i}, cond(H) = {cond(H)}")
```

```python
fig = plt.figure(figsize=(10, 10))
plt.plot(range(1, n + 1), conditions, "-o")
plt.xlabel("n")
plt.ylabel("cond(H)")
plt.title("Condition number of Hilbert matrix")
plt.yscale("log")
plt.show()
```
